# 타이타닉 데이터를 통한 생존자 유추

## 1. 기존 실습

```python
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical
from keras.models import Sequential
from keras import layers
from keras import optimizers
```

```python
path = '../9.데이터/titanic/'
```

```python
data_train = pd.read_csv(path + 'train.csv')
data_test = pd.read_csv(path + 'test.csv')
```

### 1) 학습 데이터 정규화

```python
# 학습데이터 배열 생성

data_train_np = np.zeros([891, 8])
```

```python
# 성별 데이터 원-핫 인코딩

sex_num = np.zeros(891)
cnt = 0
for d in data_train['Sex']:
    if d == 'female':
        sex_num[cnt] = 1
    cnt = cnt + 1
    
data_train_np[:, :2] = to_categorical(sex_num)
```

```python
# 좌석등급, 나이, 친인척, 가족 데이터 정규화

data_train_np[:, 2:5] = to_categorical(data_train['Pclass'].to_numpy() - 1)
data_train_np[:,5] = data_train['Age'] / 80
data_train_np[:,6] = data_train['SibSp'] / 10
data_train_np[:,7] = data_train['Parch'] / 10
```

```python
# 결측데이터 처리

data_train_np[np.isnan(data_train_np)] = 30/50
```

```python
data_train_np_y = to_categorical(data_train['Survived'])
```

### 2) 테스트 데이터 정규화

```python
# 테스트 데이터 배열 생성

data_test_np = np.zeros([418, 8])
```

```python
# 성별 데이터 원-핫 인코딩

sex_num = np.zeros(418)
cnt = 0
for d in data_test['Sex']:
    if d == 'female':
        sex_num[cnt] = 1
    cnt = cnt + 1
    
data_test_np[:, :2] = to_categorical(sex_num)
```

```python
# 좌석등급, 나이, 친인척, 가족 데이터 정규화

data_test_np[:, 2:5] = to_categorical(data_test['Pclass'].to_numpy() - 1)
data_test_np[:,5] = data_test['Age'] / 80
data_test_np[:,6] = data_test['SibSp'] / 10
data_test_np[:,7] = data_test['Parch'] / 10
```

```python
# 결측데이터 처리

data_test_np[np.isnan(data_test_np)] = 30/50
```

### 3) 모델 생성 및 학습

```python
model = Sequential()
model.add(layers.Dense(units=8, activation='relu'))
model.add(layers.Dense(units=16, activation='relu'))
model.add(layers.Dense(units=32, activation='relu'))
model.add(layers.Dense(units=64, activation='relu'))
model.add(layers.Dense(units=2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer = optimizers.RMSprop(lr=0.001), metrics=['accuracy'])
```

```python
hist = model.fit(data_train_np, data_train_np_y, epochs = 400, batch_size = 128, validation_split=0.3)
```

```python
scores = model.evaluate(data_train_np, data_train_np_y)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

```python
# 모델 학습 과정 표시하기
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
o = model.predict(data_test_np)
answer = np.zeros([418,2])
answer[:, 0] = data_test['PassengerId']
answer[:, 1] = np.argmax(o,-1)
```

```python
df_answer = pd.DataFrame(answer)
df_answer.columns = ['PassengerId', 'Survived']
df_answer = df_answer.astype('int32')
df_answer.to_csv(path + 'test_predict_BaseExample#1.csv', index = False)
```
